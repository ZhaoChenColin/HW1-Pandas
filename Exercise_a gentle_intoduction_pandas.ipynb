 - 相关性最高的前三个字段 item_sales_level， item_collected_level, and user_gender_id, item_pv_level
 - item_sales_level: 广告商品的销量等级，int类型；取值从0开始，数值越大表示销量越大
 - item_collected_level: 广告商品被收藏次数的等级，int类型；取值从0开始，数值越大表示被收藏次数越大
 - user_gender_id：用户的预测性别编号，int类型；0表示女性用户，1表示男性用户，2表示家庭用户

```python
import datetime
```

```python
df['context_timestamp']=df['context_timestamp'].map(lambda x: datetime.datetime.fromtimestamp(x))
```

```python
df2=df.set_index(['context_timestamp']).sort_index()
```

```python
change_rate=df2.groupby('item_id').rolling('3d')['is_trade'].apply(lambda x: (x.sum()-x[-1])/(len(x)-1))
```

```python
change_rate[:14]
```

```python
change_rate=np.around(change_rate.values,decimals=4)
```

```python
df3=df.set_index(['item_id','context_timestamp']).sort_index()
```

```python
df3['change_rate%']=change_rate
```
